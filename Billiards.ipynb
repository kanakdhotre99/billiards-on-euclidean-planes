```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
```

```python
def billiards(x0,y0,theta,t=1000,xl=10,yl=10):
    xvals = [x0]
    yvals = [y0]
    xspeed = 0.1
    yspeed = 0.1*np.tan(theta)
    for i in range(t):
        if -xl<xvals[-1]<xl:
            pass
        else:
            xspeed = -xspeed
        if -yl<yvals[-1]<yl:
            pass
        else:
            yspeed = -yspeed
        xvals.append(xvals[-1]+xspeed)
        yvals.append(yvals[-1]+yspeed)
    return(xvals,yvals)
```

```python
x,y = billiards(0,0,np.sqrt(2),1000)
fig, ax = plt.subplots()
line, = ax.plot(x, y, color='k')

def animate(num, x, y, line):
    line.set_data(x[:num], y[:num])
    return(line,)

ani = FuncAnimation(fig,animate,len(x),fargs=[x, y, line],interval=2,blit=True,repeat=False)
plt.show()
```
